Cada nodo en la primer capa (layer_1) tiene una cierta cantidad de correlación con el nodo en la capa previa (layer_0). Por ejemplo, si el peso que une a la primer nuerona de entrada con la primer neurona de la capa uno (laye_1), vale **1.0**, significa que la neurona de entrada influye en un **100%** sobre la primer neurona de la capa 1 y así sucesivamente, si el peso que une a estas dos neuronas valiese **1.3**, entonces la neurona de entrada influye en un **130%** sobre la primer neurona de la capa 1. Lo anterior, siempre y cuando los pesos restantes conectados a la primer neurona de la capa 1 valgan **0**. Si todos los pesos adiquieren distintos valores, la correlación que ejercen sobre la primer neurona se dividira entre estos.

**A continuación mostraremos "por qué" razón requirimos que nuestra red neuronal sea NO-LINEAL (Hasta el momento hemos hecho uso de redes neuronales LINEALES):**

Tenemos las siguientes operaciones:  

a) **5 * 10 * 2 = 100 y 5 * 20 = 100**  
b) **1 * 0.25 * 0.9 = 0.225 y 1 * 0.225 = 0.225**  

Como se observa, en el inciso **a** ambas operaciones dan como resultado el número 100. Lo mismo ocurre con el inciso **b**.  

Veamos las siguientes imagenes y veamos el porque de estas operaciones:

```python
red_1 = np.asarray(Image.open("red_multicapa_elementos.PNG").convert("RGB"))
red_2 = np.asarray(Image.open("red_tipo_perceptron.PNG").convert("RGB"))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
axes[0].axis("off")
axes[1].axis("off")

axes[0].imshow(red_1)
axes[1].imshow(red_2)

plt.show()
```

*Para la primera imagen:*  

- La segunda neurona de la capa 0 tiene 2 entradas (1.0 y -1.0). Si multiplicamos 1.0 por el peso que une a esta neurona con la segunda neurona de la capa 1 (0.25), obtendremos lo siguiente: 1.0 * 0.25 = 0.25. Si a este resultado lo multiplicamos por 0.9 (que es el peso que une a la segunda neurona de la capa 1 con la única neurona de la capa de salida), obtenemos lo siguiente: **1.0 * 0.25 * 0.9 = 0.225**. 

*Para la segunda imagen:*  

- La segunda neurona de la capa 0 tiene 2 entradas (1.0 y -1.0). Si multiplicamos la primer entrada (1.0), por el peso que une a esta neurona con la única neurona de la capa de salida obtenemos lo siguiente: **1.0 * 0.225 = 0.225**.  

Para ambas arquitecturas de red, si deseamos los procedimientos anteriores con la entrada restante (-1.0) nos daremos cuenta que el resultado de estas dos multiplicaciones nos darán como resultado **-0.225**. Ambos tipos de operaciones (tanto cuando la entrada es 1.0, como cuando es -1.0), arrojan el mismo resultado respectivamente. La única diferencia radica en la cantidad de operaciones que se realizan, mientras que para la primer red neuronal se deben realizar 3 multiplicaciones para obtener 0.255 o -0.255 (dependiendo cual sea la entrada), para la segunda red neuronal sólo se realizan 2 multiplicaciones y se obtiene le mismo resultado.  

**Lo anterior se traduce en que nuestra red neuronal derecha (la segunda imagen), obtiene exactamente el mismo resultado que la red neuronal izquierda (la primer imagen), además de ser esta última menos costosa al procesarse debido a que sólo se realiza un producto punto entre las neuronas de entrada y la neurona de salida. Por lo tanto, no tiene sentido crear una arquitectura de red "más compleja" añadiendo capas, si una red de dos capas (tipo perceptrón) puede realizar la misma tarea.**

<font color="darkorange">**En conclusión, que nuestras redes neuronales sean lineales hace referencia a que la gran cantidad de multiplicaciones pueden ser sintetizada a una sóla (como se aprecia en los ejemplos anteriores), por lo tanto, no tiene sentido incrementar la "complejidad" de nuestra red neuronal agregando capas, si esta puede ser representada por una red neuronal tipo perceptrón (como la que se aprecia en la imagen derecha).**</font> <font color="blue"> **Es aquí cuando se hace necesario el uso las funciones de activación, para darle NO-LINEALIDAD a nuestras arquitecturas de red y por ende, puedan clasificar conjuntos de datos más complejos.**</font>
